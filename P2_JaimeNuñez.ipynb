```python
#Jaime Núñez 201730031
#Juan Camilo 201812463

#Lista de imports
import time
import numpy as np
import sympy as sp
import math as mt


def capicua_num(entero):
    if entero==int(str(entero)[::-1]) and entero>=10:
        resultado=1
    else:
        resultado=0
                
    return resultado

ListaEnteros=np.array([21212,2332332,12345678987654321,15555551])

for i in ListaEnteros:
    print("El número " + str(i) + " retorna: " + str(capicua_num(i)))
```

```python
def cantidad_ncum(limite):  
    cantidad=0
    
    #Cálculo de numeros capicúas
    for i in range(limite):
        if capicua_num(i)==1:
            cantidad+=1
    return cantidad

limite=10000000
start=time.time()
cantidad=cantidad_ncum(limite)
end=time.time()
print("La cantidad de numeros capicúa entre 0 y " + str(limite) + " es: " + str(cantidad))
print("El tiempo de ejecución fue de: " + str(end-start) + " [s]")
```

**Punto 5**

En este caso se tienen los siguientes puntos sobre el plano cartesiano (x,y): (1, 3), (2, 4) y (3, 7). Por lo que se busca plantear un sistema de ecuaciones para determinar el polinomio cuadrático que interpola los puntos y hallar la solución al sistema




```python

#Planteamiento de las ecuaciones en su forma matricial
A= np.array([[1,1,1],[1,2,4],[1,3,9]])

#resultados de la interpolación
B= np.array([3, 4, 7])

#Solución del sistema para extaer los coeficientes
Res=np.linalg.solve(A,B)

print("Los coeficientes son: " + str(Res))
print("En donde el coeficiente a0 es: " + str(Res[0]))
print("mientras que el coeficiente a1 es: " + str(Res[1]))
print("y, por último, el coeficiente a2 es: " + str(Res[2]))

```
